# Anchor

```python
import cv2
import numpy as np
qs = [60,65,70,75,80,85,90, 95,100]
q2i = {60:0, 65:1, 70:2, 75:3,80:4,85:5,90:6,95:7,100:8}
kodak = []
for i in range(1,26):
    if i >= 10:
        kodak.append(f'{i}')
    else: kodak.append('0' + f'{i}')
        
anchor_rates = []
```

```python
for q in qs:
    temp = []
    for k in kodak:
        png = cv2.imread('/Users/lukasecilmis/Desktop/Kodak/kodim' + k + '.png')
        y, x, _ = png.shape
        pixels = x * y
        
        PATH = '/Users/lukasecilmis/Desktop/JPEG_DNA_Transcoder/Publication/JPEG/' + f'{q}' + '_' + k + '.jpg'
        cv2.imwrite(PATH, png, [int(cv2.IMWRITE_JPEG_QUALITY), q])
        jpg = cv2.imread(PATH)
        
        PATH2 = '/Users/lukasecilmis/Desktop/xx.png'
        cv2.imwrite(PATH2, jpg)
        png_decoded = cv2.imread(PATH2)
        
        i = open(PATH, 'rb').read()
        length = 8 * len(i)
        rate = length / pixels  
        temp.append((q, k, rate, cv2.PSNR(png, png_decoded))) 
    anchor_rates.append(temp)
```

```python
# Processing
rates = anchor_rates
anchor = []
```

```python
for q in qs:
    data = rates[q2i[q]]
    n = len(data)
    r = 0
    p = 0
    for d in data:
        r += d[2]
        p += d[3]
    anchor.append((r/n, p/n))
print(anchor)
```

```python
import matplotlib.pyplot as plt
plt.plot(*zip(*anchor))
plt.title('Anchor')
plt.xlabel('Rate (nts/pixel)')
plt.ylabel('PSNR')
plt.show()
```
